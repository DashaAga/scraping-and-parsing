```python
import requests
from bs4 import BeautifulSoup
from lxml import etree
```

```python
city = "moscow"
adults = 2
checkinDate = "2024-10-02"
checkoutDate = "2024-10-03"

url = f"https://travel.yandex.ru/hotels/{city}/?adults={adults}&bbox=37.39714713844697%2C55.62528816310007~37.78280377149642%2C55.856245&checkinDate={checkinDate}&checkoutDate={checkoutDate}&childrenAges=&geoId=213&lastSearchTimeMarker=1727858769239&navigationToken=0&oneNightChecked=false&searchPagePollingId=370b877dc38183a8261731322b1096-0-newsearch&selectedSortId=relevant-first"

response = requests.get(url)
```

```python
# Проверяем статус ответа
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # Находим все блоки с отелями
    hotel_blocks = soup.find_all('li', class_='L8lvE')  # Элемент <li> с классом 'L8lvE' содержит информацию об отелях

    hotels_data = []

    for hotel in hotel_blocks:
        # Находим название отеля
        hotel_name = hotel.find('section', class_='mPDLV').get('itemscope')
        hotel_link = hotel.find('a', class_='Kw61r').get('href')

        # Извлекаем цену
        price_tag = hotel.find('div', class_='EhCXF fBoI3 _274Q5 _4ZiVK')
        if price_tag:
            price = price_tag.find('span', class_='bQcBE PotX_').get_text().strip()
        else:
            price = 'Цена не указана'

        # Собираем данные в словарь
        hotel_info = {
            'name': hotel_name,
            'link': f"https://travel.yandex.ru{hotel_link}",
            'price': price
        }
        
        hotels_data.append(hotel_info)

    # Выводим полученные данные
    for hotel in hotels_data:
        print(f"Название: {hotel['name']}")
        print(f"Ссылка: {hotel['link']}")
        print(f"Цена: {hotel['price']}")
        print("-" * 30)
else:
    print(f"Ошибка: {response.status_code}")
```

```python
data_list = []

for i in range(1, 10):

    url = f"https://101hotels.com/main/cities/{city}?in={checkinDate}&out={checkoutDate}&adults={adults}&page={i}"
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Поиск всех блоков с информацией об отелях
        hotels = soup.find_all('article', class_='item-inner clearfix')
        
        # Извлечение информации из каждого отеля
        for hotel in hotels:
            # Название отеля
            hotel_name = hotel.find('span', itemprop='name').get_text(strip=True)

            # Извлечение URL отеля и региона (например, можно сделать словарь регионов)
            hotel_url = hotel.find('a', class_='has_query_params')['href']

            # Цена
            price_tag = hotel.find('span', class_='price-value')
            if price_tag:
                price = price_tag.get_text(strip=True).replace(' ', '')  # Убираем пробелы в цене
                price = int(price)  # Приводим цену к числовому типу
            else:
                price = None

            # Собираем данные в нужный формат
            data = {
                "checkin": checkinDate,
                "checkout": checkoutDate,
                "guests": [
                    {
                        "adults": adults,
                        "children": []
                    }
                ],
                "hotels_name": hotel_name,
                "region": city,
                "Price": price
            }

            data_list.append(data)
    else:
        print(f"Ошибка при запросе: {response.status_code}")
```

```python
for hotel_data in data_list:
    print(hotel_data)
```
